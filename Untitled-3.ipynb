```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
mall_df = pd.read_csv('Mall_customers.xls')
mall_df.head()
```

```python
X = mall_df.iloc[:,3:5]
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dend = sch.dendrogram(sch.linkage(X_scaled, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=4, linkage='ward')
labels = model.fit_predict(X_scaled)
labels
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=model.labels_, cmap='plasma')
```

```python
#principlal component analysis
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

```python
print(type(cancer))
```

```python
print(cancer['DESCR'])
```

```python
cancer_df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
cancer_df.head()
```

```python
cancer_df.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_cancer_scaled = scaler.fit_transform(cancer_df)
X_cancer_scaled
```

```python
from sklearn.decomposition import PCA
pca_model = PCA(n_components=2)
X_new = pca_model.fit_transform(X_cancer_scaled)
```

```python
X_new.shape
```

```python
X_cancer_scaled.shape
```

```python
plt.scatter(X_new[:,0],X_new[:,1],c = cancer['target'],cmap='plasma')
plt.title('PCA of Breast Cancer Dataset')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
plt.show()
```
